‚ùóÔ∏è Use the scoring metric relevant for the task



```python
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
```

```python
X = data.drop(columns='malignant')
y = data['malignant']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=0)

# Combine preprocessor and linear model in pipeline
from sklearn.linear_model import LogisticRegression

final_pipe = Pipeline([('imputer', KNNImputer()), 
                       ('scaler', MinMaxScaler()),
                       ('logistic_regression', LogisticRegression())])

from sklearn import set_config; set_config(display='diagram')

final_pipe
```

```python
from sklearn.model_selection import GridSearchCV

# Grid search KNNImputer parameter n_neighbors
grid_search = GridSearchCV(final_pipe, param_grid={"imputer__n_neighbors": [2,5,10]}, cv=5, scoring="recall")

grid_search.fit(X_train, y_train)

tuned_pipe = grid_search.best_estimator_
tuned_pipe
```

üëá What is the performance of the optimal pipeline? Make sure you cross validate!

```python
# train the tuned pipe
tuned_pipe_trained = tuned_pipe.fit(X_train,y_train)
```

üëá Using your optimal pipeline, predict wether the following tumor is malignant or not

```python
new_data = pd.read_csv("new_data.csv")
new_data

# Make predictions
malignant = tuned_pipe_trained.predict(new_data)[0]
malignant
```
